```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as datasets
from sklearn.decomposition import PCA
from sklearn.decomposition import KernelPCA
from sklearn.preprocessing import StandardScaler
```

# Makemoons

```python
X, y = datasets.make_moons(n_samples=500, noise=0.02, random_state=417)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y)
plt.show()
```

```python
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X)

plt.title("PCA")
plt.scatter(X_pca[:, 0], X_pca[:, 1], c = y)
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.show()
```

```python
kpca = KernelPCA(kernel ='rbf', gamma = 30)
X_kpca = kpca.fit_transform(X)

plt.title("Kernel PCA")
plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c = y)
plt.show()
```

# Iris

```python
df_iris = sns.load_dataset('iris')
df_iris['species'] = df_iris['species'].map({name: i for i, name in enumerate(df_iris['species'].unique())})
X = df_iris.drop(columns=['species']).values
y = df_iris['species'].values
```

```python
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X)

plt.title("PCA")
plt.scatter(X_pca[:, 0], X_pca[:, 1], c = y)
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.show()
```

```python
kpca = KernelPCA(kernel ='poly', gamma=10)
X_kpca = kpca.fit_transform(X)

plt.title("Kernel PCA")
plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c = y)
plt.legend(df_iris['species'].unique(), loc='upper right')
plt.show()
```

# Cancer

```python
df_cancer = datasets.load_breast_cancer(as_frame=True).frame
X = df_cancer.drop(columns=['target']).values
y = df_cancer['target'].values
X = StandardScaler().fit_transform(X)
```

```python
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X)

plt.title("PCA")
plt.scatter(X_pca[:, 0], X_pca[:, 1], c = y)
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.show()
```

```python
kpca = KernelPCA(kernel ='cosine', gamma=2)
X_kpca = kpca.fit_transform(X)

plt.title("Kernel PCA")
plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c = y)
plt.show()
```

# Penguins

```python
df = sns.load_dataset("penguins")
df.rename(columns={"bill_length_mm": "culmen_length_mm", "bill_depth_mm": "culmen_depth_mm"}, inplace=True)
value_columns = ["culmen_length_mm", "culmen_depth_mm", "flipper_length_mm", "body_mass_g"]
df["species"] = df["species"].map({name: i for i, name in enumerate(df["species"].unique())})
df.dropna(inplace=True)
X = df[value_columns].values
y = df["species"].values
X = StandardScaler().fit_transform(X)
```

```python
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X)

plt.title("PCA")
plt.scatter(X_pca[:, 0], X_pca[:, 1], c = y)
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.show()
```

```python
kpca = KernelPCA(kernel ='poly', gamma=10)
X_kpca = kpca.fit_transform(X)

plt.title("Kernel PCA")
plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c = y)
plt.show()
```

# AMP

I am sampling the dataset because of the size of the dataset

```python
df = pd.read_csv('../list_1/data/final_data.csv')
df = df.sample(frac=0.1, random_state=42)
values = ['Length', 'Molecular_Weight', 'Isoelectric_Point', 'Hydrophobicity', 'Instability_Index']
y = 'Classification'
X = df[values].values
y = df[y].values
X = StandardScaler().fit_transform(X)
```

```python
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X)

plt.title("PCA")
plt.scatter(X_pca[:, 0], X_pca[:, 1], c = y)
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.show()
```

```python
kpca = KernelPCA(kernel ='rbf', gamma=10)
X_kpca = kpca.fit_transform(X)

plt.title("Kernel PCA")
plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c = y, alpha=0.2, s=5)
plt.show()
```
